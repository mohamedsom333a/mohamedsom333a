```python
from numpy.random.mtrand import random_sample
from scipy.stats import skewnorm
import pandas as pd
df=pd.DataFrame(normalized_arr1[0])
a=df.skew()
print(a)
r = skewnorm.rvs(a , size=2648)  #Skewnorm function
r = r - min(r)      #Shift the set so the minimum value is equal to zero.
random_skewnorm_temp=preprocessing.normalize([np.array(r)])
f = Fitter(random_skewnorm_temp,
          distributions= 'skewnorm')
f.fit()
f.summary()
```
